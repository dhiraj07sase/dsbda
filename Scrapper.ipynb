```python
import requests
import bs4
```

```python
request1= requests.get('https://www.flipkart.com/motorola-edge-40-neo-black-beauty-128-gb/p/itm01cc46d96a79f?pid=MOBGQFX6JGTWFSGD&lid=LSTMOBGQFX6JGTWFSGDTXTCZS&marketplace=FLIPKART&store=tyy%2F4io&srno=b_1_1&otracker=browse&iid=0000a439-89ed-43bb-9ffb-31b782186fb3.MOBGQFX6JGTWFSGD.SEARCH&ssid=51lmd0ye7k0000001713167896903%27)')
request1
```

```python
request1.content
```

```python
soup=bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews=soup.findAll('div',{'class' :'ZmyHeo'});
for review in reviews:
    print(review.get_text()+ "\n\n")
```

```python
ratings=soup.find('div',{'class': 'ipqd2A'}).get_text();
print(ratings)
```

```python
individual_ratings=soup.findAll('div',{'class' :'XQDdHH Ga3i8K'});
for indi_rating in individual_ratings:
    print(indi_rating.get_text()+"\n\n")
```

```python
tags=soup.find('span', {'class' :'_6EBuvT OHdLLG'}).get_text();
tags
```

```python
customer_name= soup.findAll('p' , { 'class' : '_2NsDsF AwS1CA'});
for cust in customer_name:
    print(cust.get_text() + "\n\n")
```

```python
price= soup.find('div', {'class' : 'Nx9bqj CxhGGd'}).get_text();
print(price)
```

```python
Quetions = soup.findAll('div', {'class' : 'wys2hv _43gOsC'});
for que in Quetions :
    print(que.get_text())
    ans=que.findNext('div',{'class' :'JxAXcP'}).get_text();
    print(ans + "\n\n")
```

```python
specification = soup.findAll('td' , {'class' : '+fFi1w col col-3-12'});
for spe in specification:
    print(spe.get_text())
    ans= spe.findNext('td', {'class' : 'Izz52n col col-9-12'}).get_text();
    print(ans +"\n\n")
```
